```julia
x = linspace(0,2*pi,1000);
```

This is the Unitary version of DFT, so that output has same "length" as input.
https://en.wikipedia.org/wiki/Discrete_Fourier_transform#The_unitary_DFT

```julia
dft(A) = fft(A) / sqrt(length(A))
```

```julia
idft(A) = ifft(A) * sqrt(length(A))
```

```julia
y=[ (Float64)(a^2+cos(10a)) for a in x];
```

```julia
plot(x=x, y=y, Geom.line)
```

```julia
L(k) = [ (a < k || (2*pi -a) < k) ? 1 : 0 for a in x]
```

```julia
filter(H, S) = [ H[i] * S[i] for i in 1:length(S)]
```

```julia
plot(x=x, y=real(idft(filter(L(0.05), dft(y)))), Geom.line)
```

```julia
plot(x = x, y=log(abs(dft(y))), Geom.line)
```
